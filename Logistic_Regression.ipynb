**1) What is Logistic Regression, and how does it differ from Linear Regression ?**

Logistic Regression is a machine learning algorithm used for classification problems. It predicts probabilities and outputs values between 0 and 1, often used for binary classification (e.g., spam or not spam).

**Difference from Linear Regression:-**

**1. Purpose:-**

Linear Regression is for prediction of continuous values (e.g., house prices).

Logistic Regression is for classification (e.g., yes/no, 0/1).

**2. Output:-**

Linear Regression gives any real number.

Logistic Regression gives probabilities (0 to 1) using the sigmoid function.


**2) What is the mathematical equation of Logistic Regression.**

**The mathematical equation of Logistic Regression is:-**

$$ P(y) = \frac{1}{1 + e^{-(b_0 + b_1x_1 + b_2x_2 + ... + b_nx_n)}} $$


**3) Why do we use the Sigmoid function in Logistic Regression.**

**The Sigmoid function in Logistic Regression is used to:-**

Convert output into probabilities (0 to 1).

Handle binary classification problems.

Make the model non-linear for better decision boundaries.

**6) Explain the difference between Lasso, Ridge, and Elastic Net regression?**

- **Lasso Regression (L1 Regularization)**:  
  $$ J(\theta) = \text{Loss} + \lambda \sum |w_i| $$  

- **Ridge Regression (L2 Regularization)**:  
  $$ J(\theta) = \text{Loss} + \lambda \sum w_i^2 $$  

- **Elastic Net Regression (L1 + L2 Regularization)**:  
  $$ J(\theta) = \text{Loss} + \lambda_1 \sum |w_i| + \lambda_2 \sum w_i^2 $$


**Lasso =** Feature selection (some weights become zero).

**Ridge =** Shrinks weights but keeps all features.

**Elastic Net =** Combination of both (best of both worlds).

**7) When should we use Elastic Net instead of Lasso or Ridge ?**

**Use Elastic Net when:-**

Features are highly correlated (better than Lasso).

Lasso drops too many features (Elastic Net keeps balance).

Ridge keeps all features (Elastic Net adds feature selection).

High-dimensional data (many features, fewer samples).

**8) What is the impact of the regularization parameter (Œª) in Logistic Regression ?**

The regularization parameter (Œª) in Logistic Regression controls the strength of penalty on large coefficients:

**High Œª (Strong Regularization) ‚Üí** Shrinks coefficients more, reduces overfitting, but may lead to underfitting.

**Low Œª (Weak Regularization) ‚Üí** Shrinks coefficients less, fits data better but may overfit.

**Œª = 0 (No Regularization) ‚Üí** Works like standard Logistic Regression, increasing the risk of overfitting.

**9) What are the key assumptions of Logistic Regression?**

**Key Assumptions of Logistic Regression**

Linear relationship between independent variables and log-odds.

Independent observations (no autocorrelation).

No multicollinearity (independent variables should not be highly correlated).

Large sample size for stable estimates.

Binary or categorical dependent variable (for standard Logistic Regression).

**11) What are Classification Evaluation Metrics ?**

## **Classification Evaluation Metrics**

1. **Accuracy**  
   $$ Accuracy = \frac{TP + TN}{TP + TN + FP + FN} $$

2. **Precision**  
   $$ Precision = \frac{TP}{TP + FP} $$

3. **Recall (Sensitivity)**  
   $$ Recall = \frac{TP}{TP + FN} $$

4. **F1-Score**  
   $$ F1 = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}} $$

5. **ROC-AUC Score**  
   - Measures model‚Äôs ability to separate classes.  

6. **Confusion Matrix**  
   - Displays **TP, TN, FP, FN** for detailed analysis.  


 **12 How does class imbalance affect Logistic Regression ?**

Impact of Class Imbalance on Logistic Regression

**Biased Predictions ‚Üí** Favors majority class.

**Poor Recall ‚Üí** Misses minority class.

**Misleading Accuracy ‚Üí** High accuracy but bad classification.

**Skewed Decision Boundary ‚Üí** Wrong classification results.

Solutions

 Resampling (Oversampling/Undersampling)

 Class Weights (Give more weight to minority class)

 Use F1-score, Precision-Recall instead of Accuracy

**15) How is Logistic Regression extended for multiclass classification?**

**Extending Logistic Regression for Multiclass Classification**

Logistic Regression is extended using:

One-vs-Rest (OvR) ‚Äì Trains a separate binary classifier for each class.

One-vs-One (OvO) ‚Äì Trains classifiers for every pair of classes.

Softmax Regression (Multinomial) ‚Äì Uses the softmax function to handle all classes at once.

**Which One to Use?**

‚úî Default in Scikit-Learn ‚Üí OvR (good for most cases).

‚úî Multinomial (Softmax) ‚Üí Best for balanced multiclass problems.

**20) How do we interpret coefficients in Logistic Regression?**

Interpreting Coefficients in Logistic Regression

Each coefficient (Œ≤) represents the impact of a feature on the log-odds of the target class.

Positive Œ≤ ‚Üí Increases probability of the event.

Negative Œ≤ ‚Üí Decreases probability of the event.

Exp(Œ≤) (Odds Ratio) ‚Üí A unit increase in the feature multiplies the odds by Exp(Œ≤).

 Example: If Œ≤ = 0.5, then Exp(0.5) ‚âà 1.65, meaning the odds increase by 65%.

```python
df = df[['Pclass', 'Sex', 'Age', 'Survived']].copy()
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
df['Age'] = df['Age'].fillna(df['Age'].mean())
```
